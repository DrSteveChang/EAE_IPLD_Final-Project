```python
# The library you have to use
import numpy as np

# Some extra libraries to deal with images and files
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img_path = "../data/starry_night.png"

with Image.open(img_path) as img:
    img_arr = np.array(img)

plt.imshow(img_arr)
```

```python
# Ex 1.1: Get the height and width of the image in pixels

max_height = img_arr.shape[0]   # The first element is the height (rows)
max_width = img_arr.shape[1]    # The second element is the width (columns)

print(f"Height: {max_height} pixels")
print(f"Width: {max_width} pixels")
```

```python
# Ex 1.2: Define the dimensions of the crop, make sure that min is smaller than max and also that min is larger than 0 
# and max is smaller than the size of the image for every dimension. If not, print a message explaining the error.

# Example crop dimensions, modify them to test your code
crop_min_h = 600 
crop_max_h = 1000
crop_min_w = 550
crop_max_w = 1100


# Constraint checks
# TODO: Develop the code to validate the constraints and print the error message if any constraint is invalid
```

```python
# Constraint checks
valid_height = (0 <= crop_min_h < crop_max_h <= max_height)
valid_width = (0 <= crop_min_w < crop_max_w <= max_width)

if not valid_height:
    print(f"Error: Height dimensions {crop_min_h}:{crop_max_h} are invalid. (Image Height: {max_height})")

elif not valid_width:
    print(f"Error: Width dimensions {crop_min_w}:{crop_max_w} are invalid. (Image Width: {max_width})")

else:
    print("Dimensions are valid.")
```
