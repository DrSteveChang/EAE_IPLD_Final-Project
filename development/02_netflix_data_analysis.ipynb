```python
# Fill the NaN (missing) values with the string "Unknown" first.
movies_df['country'] = movies_df['country'].fillna('Unknown')

# Join with ", " all the elements of the list into a single string
all_countries_string = ", ".join(movies_df['country'])

# Split it by ", " to get a list of all the individual countries.
all_countries_list = all_countries_string.split(', ')

# Get the number of unique countries
n_countries = len(set(all_countries_list))

print(f"There are {n_countries} different countries in the data")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data 
movies_df = pd.read_csv(data_path, index_col='show_id')

# Filter for Movies
movies_only_df = movies_df[movies_df['type'] == 'Movie'].copy()

# Extract minutes from duration
# The format is typically "X min". We split and take the first part.
# We use .copy() to avoid SettingWithCopyWarning
movies_only_df['duration_minutes'] = movies_only_df['duration'].apply(lambda x: int(str(x).split(' ')[0]) if isinstance(x, str) else 0)

# Group by release_year and calculate mean duration
movies_avg_duration_per_year = movies_only_df.groupby('release_year')['duration_minutes'].mean()

# Plotting
fig = plt.figure(figsize=(9, 6))
plt.plot(movies_avg_duration_per_year.index, movies_avg_duration_per_year.values)
plt.title("Average Duration of Movies Across Years")
plt.xlabel("Year")
plt.ylabel("Average Duration (minutes)")
plt.grid(True)

# Save the plot
plt.savefig('avg_movie_duration_per_year.png')

# Display the series
print(movies_avg_duration_per_year)
```
