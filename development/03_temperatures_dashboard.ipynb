```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data_path = "../data/cities_temperatures.csv"
temps_df = pd.read_csv(data_path)

# Preprocessing
temps_df["Date"] = pd.to_datetime(temps_df["Date"]).dt.date
temps_df["AvgTemperatureCelsius"] = (temps_df["AvgTemperatureFahrenheit"] - 32) * 5/9

# Ex 3.6
city = "Munich"
start_date = pd.to_datetime("2008-01-01").date()
end_date = pd.to_datetime("2010-12-31").date()

# Filter by city
city_df = temps_df[temps_df["City"] == city]

# Filter by date range
# Ensure Date column is comparable (it is dt.date, start_date/end_date are dt.date)
city_df_period = city_df[(city_df["Date"] >= start_date) & (city_df["Date"] <= end_date)]

# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(city_df_period["Date"], city_df_period["AvgTemperatureCelsius"], label="Temperature")
plt.title(f"Temperature in {city} ({start_date} to {end_date})")
plt.xlabel("Date")
plt.ylabel("Avg Temperature (°C)")
plt.legend()
plt.grid(True)
plt.savefig("munich_temp_line.png")

# Histogram
plt.figure(figsize=(10, 5))
plt.hist(city_df_period["AvgTemperatureCelsius"], bins=20, edgecolor='black', label="Frequency")
plt.title(f"Temperature Distribution in {city} ({start_date} to {end_date})")
plt.xlabel("Avg Temperature (°C)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.savefig("munich_temp_hist.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and preprocess
data_path = "../data/cities_temperatures.csv"
temps_df = pd.read_csv(data_path)
temps_df["Date"] = pd.to_datetime(temps_df["Date"]).dt.date
temps_df["AvgTemperatureCelsius"] = (temps_df["AvgTemperatureFahrenheit"] - 32) * 5/9

selected_cities = ["Munich", "Buenos Aires", "Tokyo"]
start_date = pd.to_datetime("2008-01-01").date()
end_date = pd.to_datetime("2010-12-31").date()

# Line Plot
plt.figure(figsize=(15, 5))

for city in selected_cities:
    city_df = temps_df[temps_df["City"] == city]
    city_df_period = city_df[(city_df["Date"] >= start_date) & (city_df["Date"] <= end_date)]
    plt.plot(city_df_period["Date"], city_df_period["AvgTemperatureCelsius"], label=city)

plt.title(f"Temperature Comparison: {', '.join(selected_cities)} ({start_date} to {end_date})")
plt.xlabel("Date")
plt.ylabel("Average Temperature (°C)")
plt.legend()
plt.savefig("cities_comparison_line.png")

# Histogram
plt.figure(figsize=(15, 5))

for city in selected_cities:
    city_df = temps_df[temps_df["City"] == city]
    city_df_period = city_df[(city_df["Date"] >= start_date) & (city_df["Date"] <= end_date)]
    plt.hist(city_df_period["AvgTemperatureCelsius"], bins=30, alpha=0.5, label=city, edgecolor='black', linewidth=0.5)

plt.title(f"Temperature Distribution Comparison: {', '.join(selected_cities)} ({start_date} to {end_date})")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("Frequency")
plt.legend()
plt.savefig("cities_comparison_hist.png")
```
